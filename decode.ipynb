```python
  def decode_base64(self, data):
        try:
            return base64.b64decode(data).decode('utf-8')
        except Exception as e:
            self.log_error("Error during base64 decoding: {}".format(str(e)))
            raise

    def transform_data(self, df_stg):
        try:
            decode_base64_udf = udf(self.decode_base64, StringType())
            df_stg = df_stg.withColumn("decodedData", decode_base64_udf(col("data")))

            json_schema = ArrayType(StructType([
                StructField("seq", StringType()),
                StructField("rgMaterial", StringType()),
                StructField("descricaoProduto", StringType())
            ]))

            df_stg = df_stg.withColumn("jsonData", from_json(col("decodedData"), json_schema))
            return df_stg
        except Exception as e:
            self.log_error("Error during data transformation: {}".format(str(e)))
            raise
```
